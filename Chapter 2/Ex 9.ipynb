First we load the data and remove the rows with the missing values, those marked as "?".

````r
Auto = read.table("../Auto.data", header=T, na.strings="?")
print("Dimension before ommiting missing values")
dim(Auto)
print("Dimension after ommiting missing values")
Auto = na.omit(Auto)
dim(Auto)
attach(Auto)
````
