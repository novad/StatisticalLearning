## (a)

```r
library(ISLR)
attach(Auto)

lm.fit = lm(mpg~horsepower)
summary(lm.fit)
```

### i
We start by defining the null hypothesis $H_0: \beta_1 = 0$, where $\beta_1$ is the regression coefficient of the **horsepower** predictor.

Looking at the summary of the linear regression, we can see a significantly high F-statistic value, and a small p-value. This means that we can infer a relationship between the response and predictor, therefore we can reject the null hypothesis.

### ii
We see that the result deviates on average 4.906 from the true regression line (RSE), which means a percentage error of: 
$$\frac{4.906}{\bar{x}} = \frac{4.906}{23.44592} = 0.2092475$$

We see a proportion of variability in **mpg** exmplained by **horsepower** of 0.6059 ($R^2$ Statistic), that is 60.59% of the variance in the response is explained by the predictor.

### iii
The relationhip between response and predictor is negative, we see this by the negative value of $\beta_1$.
This means that **mpg** tends to be lower for higher values of **horsepower**.
### iv
Predicted **mpg** with **horsepower**=98

```r
predict(lm.fit, data.frame(horsepower = c(98)), interval = "confidence")
predict(lm.fit, data.frame(horsepower = c(98)), interval = "prediction")
```
