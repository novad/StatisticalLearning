## (a)

```r
library(ISLR)
attach(Carseats)

lm.fit = lm(Sales~Price + Urban + US)

summary(Carseats)
summary(lm.fit)
```

## (b)
**Price** seems to have a relationship with Sales, we infer this from the low p-value. Since the $\beta_Price$ is negative we see that for cars with lower prices, sales tend to be greater than for cars with higher prices.

**Urban** is a qualtivative variable indicating if the store is located in an urban or rural area. We can see from the high p-value of **UrbanYes** that sales aren't related with the location of the store.

**US** from the low p-value we can infer that the stores being located in the US tend to have higher sales. Note that in the dataset there are 1.817 times more US-based stores than non-US stores.

## (c)
The model in equation form would be:

$$y_{sales}= 13.04 + -0.05x_{price} + -0.02*Urban_{Yes} + 1.2*US_{Yes}$$



## (d)
From the F-statistic and its p-value we can infer that at least one variable is related with the response.

From the low p-values we can reject the null hypothesis of the coefficients being 0 for the variables **Price** and **USYes**

## (e)

```r
lm.fit2 = lm(Sales~Price + US)
summary(lm.fit2)
```

## (f)

From the $R^2$ statstic we see some of the variability can be explained by the regression, because of the small value the model doesn't seem to fit the data very well.

The model used in *(a)* $$y_{sales}=\hat{\beta_0}+\hat{\beta_1}x_{price}+\hat{\beta_2}x_{urban}+\hat{\beta_3}x_{us}$$ 
$$RSE = 2.472 $$ has a slightly higher than the RSE in model *(e)* $$y_{sales}=\hat{\beta_0}+\hat{\beta_1}x_{price}+\hat{\beta_2}{us}$$
$$RSE = 2.469 $$ 

## (g)
95% confidence intervals for the coefficients

```r
confint(lm.fit2)
```

```r
par(mfrow=c(2,2))
plot(lm.fit2)
```
