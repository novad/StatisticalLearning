## (a)

```r
set.seed(1)
```

```r
x = rnorm(100)
```

## (b)

```r
sd = sqrt(0.25)
eps = rnorm(100, 0, sd) ## rnorm takes the sd as argument
```

## (c)
$$Y=-1+0.5X+\epsilon$$

$\beta_0=-1$, $\beta_1=0.5$

```r
y=-1 + 0.5*x + eps

length(y)
```

## (d)

```r
plot(x, y)
```

## (e)
There's a slight difference between the estimated coefficients values for the regression and its true values.
This difference is caused by the irreducible error, in this case $\epsilon$

```r
lm.fit = lm(y~x)
coefficients(lm.fit)
```

## (f)

```r
plot(x,y)
abline(lm.fit, col="red", lwd=2)
abline(-1, 0.5, col="blue", lwd=2)

legend("topleft", legend = c("estimated reg. line", "population reg. line"), col = c("red","blue"), lwd=2)
```

## (g)

We see a small reduction in the RSE, and therefore a slight improvement of the $R^2$ statistic.
But, the high p-value for $X^2$ indicates no relationship between $X^2$ and Y, this is why we see a decrease of the F-statistic.

```r
lm.poly = lm(y~x + I(x^2))
summary(lm.fit)
summary(lm.poly)
```

## (h)

We can see a significant increase in the $R^2$ statistic, meaning that this new model fits the data better. This is obvious because of the reduced "noise"

```r
x2 = rnorm(100)
sd2 = sqrt(0.10)
eps2 = rnorm(100, 0, sd2) 

y2=-1 + 0.5*x2 + eps2

lm.fit2 = lm(y2~x2)
summary(lm.fit2)

plot(x2,y2, main="less noisy epsilon")
abline(lm.fit2, col="red", lwd=2)
abline(-1, 0.5, col="blue", lwd=2)

legend("topleft", legend = c("estimated reg. line", "population reg. line"), col = c("red","blue"), lwd=2)
```

## (i)

As expected, we see a significant decrease in the measure of the model fit, the $R^2$ statistic yields worse values.

```r
x3 = rnorm(100)
sd3 = sqrt(0.40)
eps3 = rnorm(100, 0, sd3) 

y3=-1 + 0.5*x3 + eps3

lm.fit3 = lm(y3~x3)
summary(lm.fit3)

plot(x3,y3, main="noiser epsilon")
abline(lm.fit3, col="red", lwd=2)
abline(-1, 0.5, col="blue", lwd=2)

legend("topleft", legend = c("estimated reg. line", "population reg. line"), col = c("red","blue"), lwd=2)
```

## (j)
We can see that the noisier fit has a wider interval with respect to the ones with less noise.

```r
confint(lm.fit)
confint(lm.fit2)
confint(lm.fit3)
```
