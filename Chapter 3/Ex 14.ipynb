## (a)


```r
set.seed(1)
x1 = runif(100)
x2 = 0.5*x1 + rnorm(100)/10
y = 2 + 2*x1 + 0.3*x2 + rnorm(100)
```

The regression coefficients are: $\beta_0 = 2$, $\beta_1 = 2$, $\beta_2 = 0.3$

$$Y = 2+2x_1+0.3x_2+\epsilon$$

## (b)

```r
plot(x1, x2)
cor(x1, x2)
```

## (c)

```r
lm.fit = lm(y~x1 + x2)
summary(lm.fit)
```

$\hat{\beta_0}=2.4305$, $\hat{\beta_1}=1.4396$, $\hat{\beta_3}=0.7097$

There errors between the true value of the coefficients and the estimates are significantly high.

The p-value for $\beta_1$ could be considered to be low enough to reject the null hypothesis (it's a pinch bellow the 5% cut-off). $\beta_2$ on the other hand has a value of 53%, we cannot reject the null hypothesis that $H_0: \beta_2=0$.

## (d)

```r
lm.fitx1 = lm(y~x1)
summary(lm.fitx1)
```

The p-value of the coefficient estimate is greatly reduced, therefore we can reject the null hypothesis.

## (e)

```r
lm.fitx2 = lm(y~x2)
summary(lm.fitx2)
```

From the low p-value of the coefficient we can reject the null hypothesis.

## (f)

The results do not contradict eachother, we see that there's collinearity between $x_1$ and $x_2$. 
If we compare the standard errors of the linear regressions we've made, we can see a significant reduction when using the variables on different models.

For $x_1$ we get an improvement from `0.7212`, to `0.3955`
And for $x_2$ from `1.1337` to `0.6330`

We know that the _power_ of the hypothesis test is reduced by collinearity.

In this case we can confirm the prescence of collinearity by looking at the correlation of the variables.

```r
cor(x1,x2)
```

## (g)


```r
x1 = c(x1, 0.1)
x2 = c(x2, 0.8)
y = c(y, 6)

lm.fit.all = lm(y~x1+x2)
lm.fit.x1 = lm(y~x1)
lm.fit.x2 = lm(y~x2)

summary(lm.fit.all)
summary(lm.fit.x1)
summary(lm.fit.x2)
```

From the first model we can see that $x_1$ doesn't explain much of the variability of the model now, on the other hand the p-value of $x_2$ (below 1.5%) tells us that this variable is associated with the response.



To check for outliers let's plot the studentized residuals for each model fit.
There are no outliers, except for one point in the second model.

```r
plot(predict(lm.fit.all), rstudent(lm.fit.all))
plot(predict(lm.fit.x1), rstudent(lm.fit.x1))
plot(predict(lm.fit.x2), rstudent(lm.fit.x2))
```

To check for high leverage point we check the _Residuals/Leverage_ plots.

```r
par(mfrow=c(2,2))
plot(lm.fit.all)
```

```r
par(mfrow=c(2,2))
plot(lm.fit.x1)
```

```r
par(mfrow=c(2,2))
plot(lm.fit.x2)
```

We can see that the 101th observation is a high leverage point in the first and third models.
